## Pandas
  
Código:  https://github.com/pandas-dev/pandas  
Guia: https://pandas.pydata.org/pandas-docs/stable/user_guide/index.html  
  
Questão importante, o nome... **"Panel Data"** e **"Python Data Analysis"**
  
Pandas é uma biblioteca Python, nela podemos manipular um conjunto de dados.  
  
Essa manipulacão nos permite analisar, limpar, formatar e explorar dados.

Esta biblioteca, construída a partir do Numpy, possibilita a estruturação e manipulação de dados de maneira simples e eficiente.  
  
Como os dados são a matéria prima de todo projeto de data science, manipulá-los é fundamental! Por isso, utilizaremos o Pandas em quase todas as aulas daqui pra frente!
