```python
import csv
from datetime import datetime
import json

def buscar_tweets_por_data():
    global resultados
    print("Op√ß√£o 1 - Buscar tweets por data")
    data = input("Digite a data da pesquisa (no formato dd/mm/aaaa): ") #input pedindo no formato errado
    try:
        data_pesquisa = datetime.strptime(data, '%d/%m/%Y').strftime('%Y-%m-%d') #'%Y-%m-%d'
    except ValueError:
        print("Formato de data inv√°lido. Use o formato dd/mm/aaaa.")
        return

    resultados = []

    with open('projeto.csv', 'r', encoding='utf-8') as arquivo:
        leitor_csv = csv.reader(arquivo)
        cabecalho = next(leitor_csv)
        indice_data = cabecalho.index('date')

        resultados = [linha for linha in leitor_csv if data_pesquisa in linha[indice_data]]
        print("data | conte√∫do | assunto")
        print("-" * 30)

        for linha in resultados:
        #coloquei os resultados saindo na tela com a hora formatada tb
            data = datetime.strptime(linha[0], '%Y-%m-%d %H:%M:%S%z').strftime('%d/%m/%Y')
            conteudo = linha[3]
            assunto = linha[4]
            print(f"{data} | {conteudo} | {assunto}")

def buscar_tweets_por_termo():
    global resultados
    print("Op√ß√£o 2 - Buscar tweets por termo")
    termo = input("Digite o termo da pesquisa: ").lower()
    resultados = []

    with open('projeto.csv', 'r', encoding='utf-8') as arquivo:
      leitor_csv = csv.reader(arquivo)
      cabecalho = next(leitor_csv)
      indice_conteudo = cabecalho.index('content')
      for linha in leitor_csv:
        if termo.lower() in linha[indice_conteudo].lower():
          resultados.append(linha)
          resultados = [linha for linha in leitor_csv if termo in linha[indice_conteudo].lower()]
      print("data | conte√∫do | assunto")
      print("-" * 30)

    for linha in resultados:
        data = datetime.strptime(linha[0], '%Y-%m-%d %H:%M:%S%z').strftime('%d/%m/%Y')
        conteudo = linha[3]
        assunto = linha[4]
        print(f"{data} | {conteudo} | {assunto}")

def buscar_tweets_por_assunto():
    global resultados
    print("Op√ß√£o 3 - Buscar tweets por assunto")
    while True:
        print('Pesquisar Assuntos:\n'
              '1. Copa do Mundo\n'
              '2. Elei√ß√µes\n'
              '3. Ci√™ncia de Dados\n'
              '4. Covid-19')
        entrada = input(f'Digite uma op√ß√£o para a busca em assunto: ')
        if entrada in ['1', '2', '3', '4']:
            lista_assunto = {'1': 'copa do mundo', '2': 'elei√ß√µes', '3': 'ci√™ncia de dados', '4': 'covid-19'}
            assunto = lista_assunto[entrada]
            break
        else:
            print('Digite uma op√ß√£o v√°lida.')
    resultados = []
    with open('projeto.csv', 'r', encoding='utf-8') as arquivo:
      leitor_csv = csv.reader(arquivo)
      cabecalho = next(leitor_csv)
      indice_conteudo = cabecalho.index('subject')

      resultados = [linha for linha in leitor_csv if assunto in linha[indice_conteudo].lower()]
      print("data | conte√∫do | assunto")
      print("-" * 30)

    for linha in resultados:
        data = datetime.strptime(linha[0], '%Y-%m-%d %H:%M:%S%z').strftime('%d/%m/%Y')
        conteudo = linha[3]
        assunto = linha[4]
        print(f"{data} | {conteudo} | {assunto}")

def salvar_resultado_da_busca():
    global resultados
    print("Op√ß√£o 4 - Salvar resultado da busca")

    with open('arquivo_saida.json', 'w', encoding='utf-8') as json_novo:
        cabecalho = (f'Data | Conte√∫do | Assunto')
        json_novo.write(json.dumps(cabecalho, ensure_ascii=False))
        json_novo.write("\n")
        for linha in resultados:
            item = (f'{linha[0]} | {linha[3]} | {linha[4]}')
            json_novo.write(json.dumps(item, ensure_ascii=False))
            json_novo.write(",\n")
    print("Resultados salvos em arquivo_saida.json")

def main():
    print("Boas vindas ao nosso sistema:")
    print("1 - Buscar tweets por data")
    print("2 - Buscar tweets por termo")
    print("3 - Buscar tweets por assunto")
    print("4 - Salvar resultado da busca")
    print("5 - Sair")
    while True:
        escolha = input("Escolha uma op√ß√£o (1-5): ")

        if escolha == '1':
            buscar_tweets_por_data()
        elif escolha == '2':
            buscar_tweets_por_termo()
        elif escolha == '3':
            buscar_tweets_por_assunto()
        elif escolha == '4':
            salvar_resultado_da_busca()
        elif escolha == '5':
            print("At√© logo!")
            break
        else:
            print("Op√ß√£o inv√°lida. Tente novamente.")
```

```python
main()
```
